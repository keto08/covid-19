```python
def corrplot(data, x, y, xlim=None, ylim=None, xlabel=None, ylabel=None, save=None, quantiles=False):
    df = data.copy()
    if quantiles:
        df[x] = df[x].rank(method="average", pct=True)
        df[y] = df[y].rank(method="average", pct=True)
        
    if xlim is None:
        xmin = df[x].min()
        xmax = df[x].max()
        xlim = (xmin,xmax)
    
    if ylim is None:
        ymin = df[y].min()
        ymax = df[y].max()
        ylim = (ymin,ymax)
    
    g = sns.jointplot(x, y, data=df,
                      kind="reg", truncate=False,
                      xlim=xlim, ylim=ylim,
                      height=7, space=1)
    
    if xlabel is not None:
        plt.xlabel(xlabel)
        
    if ylabel is not None:
        plt.ylabel(ylabel)
        
    r, p = stats.pearsonr(data[x],data[y])
    p_str = f'= {p:.5f}' if p >= 1E-5 else "< 0.00001"
    plt.title(f"$r^2 = {r**2:.5f}$, $p {p_str}$", y=1.3, fontsize = 16)
        
    if save is not None:
        plt.savefig(save, figsize=(8,8), dpi=300, bbox_inches="tight")
    else:
        plt.show()
```

```python
def corrfit(data, X, Y, xlabel=None, ylabel=None, log=(False, False), xlim=None, ylim=None, size=None):
    df = data.loc[:,[X,Y]]

    if log[0]:
        U = f"log({X})"
        xlabel = f"log: {xlabel}"
        df[U] = df[X].map(np.log)
    else:
        U = X

    if log[1]:
        V = f"log({Y})"
        ylabel = f"log: {ylabel}"
        df[V] = df[Y].map(np.log)
    else:
        V = Y

    r = stats.linregress(df[U], df[V])
    df['fit'] = (r.intercept + r.slope*df[U])
    df['residuals'] = df[V]  - df.fit
    df['scaled_residuals'] = df.residuals/df.fit

    dfBULK = df[df.scaled_residuals.abs() <= 2*df.scaled_residuals.std()]
    dfOUT = df[df.scaled_residuals.abs() > 2*df.scaled_residuals.std()]

    if xlim is None:
        xmin = df[U].min()
        xmax = df[U].max()
        xwidth = xmax - xmin
        xmax = xmax + 0.05*xwidth
        xmin = xmin - 0.05*xwidth
        xlim = (xmin,xmax)

    if ylim is None:
        ymin = df[V].min()
        ymax = df[V].max()
        ywidth = ymax - ymin
        ymax = ymax + 0.05*ywidth
        ymin = ymin - 0.05*ywidth
        ylim = (ymin,ymax)

    #print(size, data.columns)
    if size is not None:
        if size not in df.columns:
            if size not in data.columns:
                size = None
            else:
                df[size] = data[size]
    #print('Size:', size, df[size].min(), df[size].max())
        
    df['Outlier'] = df.scaled_residuals.abs() > 2*df.scaled_residuals.std()

    g0 = sns.jointplot(U, V, data=dfBULK,
                      kind="reg", 
                      marker="o",
                      xlim=xlim, ylim=ylim,
                      scatter=False,
                      height=7)    
    g1 = sns.scatterplot(U, V, data=df, hue="Outlier",
                         size=size, sizes=(25,200),
                         alpha=0.75, 
                         marker="o", edgecolor="k",
                         ax=g0.ax_joint)
    #g2 = sns.scatterplot(U, V, data=dfOUT, size="dPOP",
    #                    color="orange", marker="o", edgecolor="k",
    #                    ax=g0.ax_joint)

    if xlabel is not None:
        plt.xlabel(xlabel)

    if ylabel is not None:
        plt.ylabel(ylabel)

    r, p = stats.pearsonr(dfBULK[U],dfBULK[V])
    p_str = f'= {p:.5f}' if p >= 1E-5 else "< 0.00001"
    plt.title(f"$r^2 = {r**2:.5f}$, $p {p_str}$", pad=-5, fontsize = 12)
    plt.show()
    
    return stats.linregress(dfBULK[U], dfBULK[V])
```

```python
def scale(X):
    return (X - X.mean())/X.std()
```

```python
df2 = df1.loc[:,['population','dPOP','dHHS','h1','h2','h3','h4','h5','h6','h7','h8','W17','households', "0-29", "30-59", "60-89"]].copy()
df2['logPOP'] = scale(df2['population'].map(np.log))
df2['dPOP'] = scale((df2['dPOP']).map(np.log))
df2['dHHS'] = scale(df2['dHHS'])
df2['0-29'] = scale(df2['0-29'])
df2['30-59'] = scale(df2['30-59'])
df2['60-89'] = scale(df2['60-89'])
df2['d1'] = scale((df2['h1']/df2['households']).map(np.log))
df2['d2'] = scale((df2['h2']/df2['households']).map(np.log))
df2['d3'] = scale((df2['h3']/df2['households']).map(np.log))
df2['d4'] = scale((df2['h4']/df2['households']).map(np.log))
df2['d5'] = scale((df2['h5']/df2['households']).map(np.log))
df2['d6'] = scale((df2['h6']/df2['households']).map(np.log))
df2['d7'] = scale((df2['h7']/df2['households']).map(np.log))
df2['d8'] = scale((df2['h8']/df2['households']).map(np.log))

factors = ['logPOP','dPOP','dHHS','d1','d2','d3','d4','d5','d6','d7','d8',"0-29","30-59", "60-89"]
C = df2.loc[:,factors].corr().fillna(0).to_numpy()
e, v = linalg.eig(C)

print("Eigenvalues", e)
results = {}
results['components'] = factors
results['eigenvalues'] = e

components = []
for i in range(len(e)):
    df2[f'PCA{i}'] = (df2.loc[:,factors]*v[:,i]).sum(axis=1)
    
    r, p = stats.pearsonr(df2[f'PCA{i}'],df2['W17'])
    print(f"PCA{i}: ", r, p)
    if p < 0.1 or np.abs(r) > 0.1:    
        regress = corrfit(df2, X=f'PCA{i}', Y='W17', 
                        xlabel=f"PCA {i} ({','.join(factors)})", ylabel="death Week", 
                        log=(False, False))
        results[f'PCA{i}'] = {
            'eigenvector': v[:,i],
            'fit': regress
        }
        components.append(f'PCA{i}')

print(components)
df2['PCA'] = 0
weights = np.zeros_like(results['PCA0']['eigenvector'])
for pca in components:
    df2['PCA'] += results[pca]['fit'].slope*df2[pca]
    weights += results[pca]['fit'].slope*results[pca]['eigenvector']
r = corrfit(df2,'PCA','W17')
df2["PCR"] = r.intercept + r.slope*df2["PCA"]
print(r)
data = df2.copy()
pd.Series(weights, index=results['components'])
```
