```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import geopandas as gpd
```

```python
import seaborn as sns
sns.set(style="whitegrid")
```

```python
def geoplot(df, variable, title=None, cmap='YlGn', save=None):
    #variable = 'd_POP'
    umax = df[variable].max()
    umin = df[variable].min()

    fig, ax = plt.subplots(1, figsize=(8, 6), dpi=150)
    df.plot(column=variable, cmap=cmap, ax=ax, linewidth=0.125, edgecolor='0.0')
    ax.axis('off')

    sm = plt.cm.ScalarMappable(cmap=cmap, norm=plt.Normalize(vmin=umin, vmax=umax))
    sm._A = []
    cbar = fig.colorbar(sm)
    
    if title is not None:
        plt.title(title)

    if save is not None:
        plt.savefig(save, figsize=(8,6), dpi=300, bbox_inches="tight")
    else:
        plt.show()
```

```python
uk = gpd.read_file('uk_shp/6516bbf7-c7b4-451c-a8db-f6e6b5190401202045-1-1t1tsly.5n0n.shp').to_crs(epsg=3395)\
        .rename(columns={'ctyua19cd':'code', 'ctyua19nm':'name', 'st_area(sh': 'sqm','st_length(': 'length'})\
        .drop(columns=['objectid','bng_e','bng_n','ctyua19nmw'])
uk.loc[uk.code == "E10000002","code"] = "E06000060"
uk = uk.set_index('code').sort_index()
uk.head()
```

```python
geoplot(uk, 'sqm')
```

```python
df = pd.read_csv('uk_population_details.csv', skiprows=4, skipfooter=7, engine='python', thousands=',')\
        .rename(columns={'Code':'code', 'Name':'name', 'Geography1':'geography', 'All ages':'population'})\
        .set_index(['geography','code','name']).sort_index().astype(int).reset_index(['name','geography'])
df['hierarchy'] = np.nan
df.loc[df.name=='UNITED KINGDOM','hierarchy'] = 0
df.loc[df.name.isin(['ENGLAND', 'SCOTLAND', 'WALES', 'NORTHERN IRELAND']), 'hierarchy'] = 1
df.loc[df.geography.isin(['Region']), 'hierarchy'] = 2
df.loc[df.geography.isin(['County','Metropolitan County']), 'hierarchy'] = 3
df.loc[df.geography.isin(['Metropolitan District','Non-metropolitan District','Unitary Authority','London Borough','Council Area']), 'hierarchy'] = 4
df = df.dropna()
df['hierarchy'] = df.hierarchy.astype(int)
pop = df.reset_index().set_index(['hierarchy','geography','code','name']).sort_index()
pop.head()
```

```python
uk_pop = uk.join(pop.reset_index(['hierarchy','geography','name'], drop=True).population)
uk_pop['dPOP'] = (1E6*uk_pop.population/uk_pop.sqm)
geoplot(uk_pop, 'dPOP', title="Population Density, (People/km$^2$)", save="uk_pop.jpg")
```

```python
df = pd.read_csv('households.csv') #.set_index('code').drop(columns=['date','name','Rural Urban'])
df.loc[df.code.isin(['E06000028','E06000029','E10000002','E10000009']),'code'] =\
df.loc[df.code.isin(['E06000028','E06000029','E10000002','E10000009']),'code'].map({
    'E06000028':'E06000058',
    'E06000029':'E06000058',
    'E10000002':'E06000060',
    'E10000009':'E06000059'
})
df = df.groupby(['code']).sum().drop(columns=['date'])
uk_hhs = uk_pop.join(df, how="outer").dropna()
uk_hhs['dHHS'] = uk_hhs.population/uk_hhs.households
uk_hhs.head()
```

```python
geoplot(uk_hhs, 'dHHS', title="Mean Household Size", save="d_HHS.jpg")
```

```python
deaths = pd.read_csv('detailed_deaths.csv', skiprows=3, usecols=[0,3,4,5,6])\
    .rename(columns={
            'Area code': 'code', 'Geography type': 'geo_type', 'Area name ': 'name', 'Cause of death': 'cause',
            'Week number': 'week', 'Place of death': 'place', 'Number of deaths': 'deaths'
        }).set_index(['cause','code','place','week']).sort_index().unstack('place').fillna(0).astype(int)\
    .rename(columns={
            'Care home': 'care_home', 'Elsewhere': 'elsewhere', 'Home': 'home', 'Hospital': 'hospital',
            'Other communal establishment': 'other_communal', 'Hospice': 'hospice'
        })
deaths.columns = deaths.columns.droplevel()
deaths.head()
```

```python
df = deaths.groupby(['cause','code']).sum().loc[('COVID 19',)]
df['general'] = df.loc[:,['home','hospital','elsewhere']].sum(axis=1)
df['restricted'] = df.loc[:,['care_home','hospice','other_communal']].sum(axis=1)
df = df.loc[:,['general','restricted']]
```

```python
test = df.join(uk_hhs)
test = test[test.isna().any(axis=1)]
d2c = pd.read_csv('district_to_country_2019.csv').set_index('LAD19CD')
t = test.loc[test.population.isna(),['name']].join(d2c).drop(columns=['name','FID','LAD19NM','CTY19NM']).dropna()
code_map = t.CTY19CD #.to_dict()
u = df.join(code_map)
u['CD'] = u.index
u.loc[u.CTY19CD.notnull(),'CD'] = u.loc[u.CTY19CD.notnull(),'CTY19CD']
v = u.drop(columns=['CTY19CD']).groupby('CD').sum().reset_index()
v.loc[v.CD == "E10000002","CD"] = "E06000060"
v = v.rename(columns={'CD':'code'}).set_index('code')
uk_deaths = uk_hhs.join(v).dropna()
uk_deaths.head()
```

```python
uk_deaths['d_general'] = 1E5*uk_deaths.general/uk_deaths.population
uk_deaths['d_restricted'] = 1E5*uk_deaths.restricted/uk_deaths.population
uk_deaths.head()
```

```python
geoplot(uk_deaths, 'd_restricted', title="Mortality (Restricted)", cmap="copper", save="uk-deaths-restricted.jpg")
```

```python
geoplot(uk_deaths, 'd_general', title="Mortality (General)", cmap="copper", save="uk-deaths-general.jpg")
```

```python
df = uk_deaths.loc[:,['name','population','households','1', '2', '3', '4', '5', '6', '7', '8']].copy()
df.head()
```

```python
dfF = df.loc[:,['1', '2', '3', '4', '5', '6', '7', '8']].transpose().cumsum().transpose()
dfF.head()
```

```python
PHHS = 1 - dfF.divide(df.households, axis=0).shift(axis=1).fillna(0)
PHHS.head()
```

```python
kmean = {}
for i in PHHS.index:
    u = PHHS.loc[i].to_dict()
    dk = {int(k):u[k] for k in u}
    dk[9] = 0
    pk = pd.Series(dk).to_frame('Ph').reset_index().rename(columns={'index':'h'})
    pk['ph'] = pk.Ph.diff(-1) #.dropna()
    pk['k'] = pk.h
    pk['pk'] = (pk.h*pk.ph)/(pk.h*pk.ph).sum()
    pk['Pk'] = 1-pk.pk.cumsum().shift().fillna(0)
    pk = pk.dropna()
    kmean[i] = {
        'kmean': (pk.k*pk.pk).sum(),
        'k2mean': (pk.k*pk.k*pk.pk).sum()
    }
results = pd.DataFrame(kmean).transpose()
results['Tc'] = results.kmean/(results.k2mean - results.kmean)
results = results.sort_values('Tc')
```

```python
uk_results = uk_deaths.join(results)
```

```python
def corrplot(data, x, y, xlim=None, ylim=None, xlabel=None, ylabel=None, save=None, quantiles=False):
    df = data.copy()
    if quantiles:
        df[x] = df[x].rank(method="average", pct=True)
        df[y] = df[y].rank(method="average", pct=True)
        
    if xlim is None:
        xmin = df[x].min()
        xmax = df[x].max()
        xlim = (xmin,xmax)
    
    if ylim is None:
        ymin = df[y].min()
        ymax = df[y].max()
        ylim = (ymin,ymax)
    
    g = sns.jointplot(x, y, data=df,
                      kind="reg", truncate=False,
                      xlim=xlim, ylim=ylim,
                      height=7)
    
    if xlabel is not None:
        plt.xlabel(xlabel)
        
    if ylabel is not None:
        plt.ylabel(ylabel)
        
    if save is not None:
        plt.savefig(save, figsize=(8,8), dpi=300, bbox_inches="tight")
    else:
        plt.show()
```

```python
corrplot(uk_results, "Tc", "d_general", 
         xlabel="Cretical Transmissibility, $T_c$",
         ylabel="COVID-19 Mortaility (General)", 
         xlim=(0.25,0.55), ylim=(0,110), quantiles=False)
```
