```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import geopandas as gpd
```

```python
import seaborn as sns
sns.set(style="whitegrid")
```

```python
uk = gpd.read_file('uk_shp/6516bbf7-c7b4-451c-a8db-f6e6b5190401202045-1-1t1tsly.5n0n.shp').to_crs(epsg=3395)\
        .rename(columns={'ctyua19cd':'code', 'ctyua19nm':'name', 'st_area(sh': 'sqm','st_length(': 'length'})\
        .drop(columns=['objectid','bng_e','bng_n','ctyua19nmw'])\
        .set_index('code').sort_index()
uk.head()
```

```python
variable = 'length'
umax = uk[variable].max()
umin = uk[variable].min()

fig, ax = plt.subplots(1, figsize=(12, 8), dpi=100)
uk.plot(column=variable, cmap='YlGn', ax=ax, linewidth=0.5, edgecolor='0.8')
ax.axis('off')

sm = plt.cm.ScalarMappable(cmap='YlGn', norm=plt.Normalize(vmin=umin, vmax=umax))
sm._A = []
cbar = fig.colorbar(sm)

plt.show()
```

```python
df = pd.read_csv('population_2019.csv').set_index('code').rename(columns={'All Ages':'population'}).loc[:,['population']]
data = df1.join(df, how="inner")
```

```python
data['d_POP'] = 1E6*data.population/data.sqm
```

```python
data.head()
```

```python
def geoplot(df, variable, title=None, save=None):
    #variable = 'd_POP'
    umax = df[variable].max()
    umin = df[variable].min()

    fig, ax = plt.subplots(1, figsize=(8, 6), dpi=100)
    df.plot(column=variable, cmap='YlGn', ax=ax, linewidth=0.5, edgecolor='0.8')
    ax.axis('off')

    sm = plt.cm.ScalarMappable(cmap='YlGn', norm=plt.Normalize(vmin=umin, vmax=umax))
    sm._A = []
    cbar = fig.colorbar(sm)
    
    if title is not None:
        plt.title(title)

    if save is not None:
        plt.savefig(save, figsize=(8,6), dpi=300, bbox_inches="tight")
    else:
        plt.show()
```

```python
geoplot(data, 'd_POP', title="Population Density", save="d_POP.jpg")
```

```python
data['covid_death'] = data.covid_death.astype(int)
data['d_COV'] = 100000*data.covid_death/data.population
```

```python
geoplot(data, 'd_COV', title="COVID-19 Mortality", save='d_COVID.jpg')
```

```python
df = pd.read_csv('households.csv').set_index('code').drop(columns=['date','name','Rural Urban'])
u = data.join(df)
u['d_HHS'] = u.population/u.households
data = u
```

```python
geoplot(data, 'd_HHS', title="Mean Household Size", save="d_HHS.jpg")
```

```python
g = sns.jointplot("d_POP", "d_COV", data=data,
                  kind="reg", truncate=False,
                  xlim=(0,20000), ylim=(0,120),
                  height=7)
plt.xlabel('Population Density, People/km$^2$')
plt.ylabel('COVID-19 Mortality, Deaths/$10^5$')
plt.savefig('POP_COV.jpg', figsize=(8,8), dpi=300, bbox_inches="tight")
#plt.show()
```

```python
df = pd.read_csv('ethnic-population.csv').rename(columns={
        'local authority: county / unitary (prior to April 2015)': 'name',
        'Ethnicity': 'ethnicity',
        'Value': 'value',
        'Ethnicity_type': 'level'
    }).drop(columns=['Measure','Value_type','Source','Time','Geography'])
v = uk.reset_index().set_index('name').code.to_frame()
df = v.join(df.set_index('name')).dropna().reset_index().set_index(['code','ethnicity']).sort_index()
df.head()
```

```python
df[df.level=='ONS 5+1 2011'].index.unique(level=1)
```

```python
v = df[df.level=='ONS 5+1 2011'].value.unstack('ethnicity')
w = uk.loc[:,['geometry']].join(v)

index = 0
for E in df[df.level=='ONS 5+1 2011'].index.unique(level=1):
    geoplot(w, E, title=E, save=f'ETH_{index:02d}.jpg')
    index += 1
```
